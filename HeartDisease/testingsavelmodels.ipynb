```python
# Import necessary libraries for data handling and model loading
import pandas as pd
import numpy as np
import joblib
```

```python
# Define the feature names used in the original training
selected_features = [
    'age', 'gender', 'cholesterol', 'gluc', 'smoke', 'alco', 'active',
    'bmi', 'pulse_pressure', 'median_income', 'stress', 'diet_risk'
]
```

```python
def get_hardcoded_input():
    """
    Provide hardcoded input values for all required features.
    - Define a single set of feature values based on reasonable ranges.
    - Return a DataFrame with the input values.
    
    Returns:
        pd.DataFrame: DataFrame containing hardcoded input for all features
    """
    # Hardcoded input values (based on reasonable ranges from the dataset)
    input_data = {
        'age': 50,              # Age in years
        'gender': 2,            # 1: female, 2: male
        'cholesterol': 2,       # 1: normal, 2: above normal, 3: well above normal
        'gluc': 1,              # 1: normal, 2: above normal, 3: well above normal
        'smoke': 0,             # 0: non-smoker, 1: smoker
        'alco': 0,              # 0: non-drinker, 1: drinker
        'active': 1,            # 0: not active, 1: active
        'bmi': 27.5,            # Calculated as weight / (height in meters)^2
        'pulse_pressure': 50,   # Systolic - diastolic blood pressure
        'median_income': 75000, # Income in dollars
        'stress': 0.75,         # Calculated as age/100 + ap_hi/200
        'diet_risk': 0.7        # 0.3 if cholesterol=1, else 0.7
    }
    
    # Convert to DataFrame
    return pd.DataFrame([input_data], columns=selected_features)
```

```python
def predict_with_saved_models():
    """
    Load saved models and scaler from the models/ directory, use hardcoded input, and predict cardiovascular disease.
    - Load the scaler and all trained models from the models/ directory.
    - Scale the hardcoded input.
    - Make predictions with each model and display results.
    """
    # Load the saved scaler
    try:
        scaler = joblib.load('../models/scaler.pkl')
    except FileNotFoundError:
        print("Error: scaler.pkl not found. Please ensure the scaler file exists in the models/ directory.")
        return
    
    # Get hardcoded input
    input_data = get_hardcoded_input()
    
    # Scale the input data
    input_scaled = scaler.transform(input_data[selected_features])
    
    # List of model names
    model_names = ['RandomForest', 'XGBoost', 'LightGBM', 'DecisionTree', 'KNN', 'LogisticRegression']
    
    # Print the input data
    print("\nHardcoded Input Data for Prediction:")
    print(input_data.to_string(index=False))
    print("\nModel Predictions (0: No Cardiovascular Disease, 1: Cardiovascular Disease):")
    
    # Load each model and make predictions
    for name in model_names:
        try:
            model = joblib.load(f'../models/{name}_model.pkl')
            prediction = model.predict(input_scaled)[0]
            print(f"{name}: {prediction}")
        except FileNotFoundError:
            print(f"{name}: Model file not found. Please ensure '{name}_model.pkl' exists in the models/ directory.")
```

```python
if __name__ == "__main__":
    predict_with_saved_models()
```
